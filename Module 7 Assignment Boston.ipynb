### Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

**HYPOTHESIS**
- Null Hypothesis = No difference in median value of houses
- Alternative Hypothesis = Statistical difference in median value of houses near the Charles River

```python
#Boston Dataframe CHAS value counts
boston_df["CHAS"].value_counts()
```

```python
# Identifies Median value of all properties with CHAS Value of 0 (Not Near River)
a = boston_df[boston_df["CHAS"] == 0]["MEDV"]
a
```

```python
# Identifies Median value of all properties with CHAS Value of 1 (Near River) 
b = boston_df[boston_df["CHAS"] == 1]["MEDV"]
b
```

```python
scipy.stats.ttest_ind(a,b,axis=0,equal_var=True)
```

**FINDING: pvalue is greater than 0.05. Fail to reject the null hypothesis since there is NO statistical significance.** 

### Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

**HYPOTHESIS**
- Null Hypothesis = No difference in Median Value of houses built prior to 1940
- Alternative Hypothesis = Statistical difference in Median value of houses built prioer to 1940

```python
#Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)
from statsmodels.formula.api import ols
lm = ols('MEDV ~ AGE', data = boston_df).fit()
table = sm.stats.anova_lm(lm)
print(table)
```

**FINDING: There is a statistical difference in the median vlaue of houses**
